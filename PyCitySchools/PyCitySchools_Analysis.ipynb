```python
#python dependencies 
import pandas as pd
from pathlib import Path
```

```python
#Path to obatian the data
school_data_path= Path(r"C:\Users\jrobe\Documents\TDM-VIRT-DATA-PT-02-2024-U-LOLC\PandasChallenge\Pandas-Challenge\Resources\schools_complete.csv")
student_data_path = Path(r"C:\Users\jrobe\Documents\TDM-VIRT-DATA-PT-02-2024-U-LOLC\PandasChallenge\Pandas-Challenge\Resources\students_complete.csv")
#Data reading
schoolData = pd.read_csv(school_data_path)
studentData = pd.read_csv(student_data_path) 

district_summary = pd.merge(schoolData, studentData, how = "left", on=["school_name", "school_name"])
district_summary.head()
```

```python
total_unique_schools = district_summary['school_name'].nunique()
Total_students = district_summary['student_name'].count()
Total_budget = district_summary['budget'].unique().sum()
Average_Math_score = district_summary['math_score'].mean()
Average_Reading_score = district_summary['reading_score'].mean()
passing_Math_count = district_summary[district_summary['math_score'] >= 70]['student_name'].count()
passing_Reading_count = district_summary[district_summary['reading_score'] >= 70]['student_name'].count()
passing_math_reading_count = district_summary[(district_summary['math_score'] >= 70) & (district_summary['reading_score'] >= 70)]['student_name'].count()
print(f"Total Unique schools = {total_unique_schools}")
print(f"Total Students = {Total_students}")
print(f"Total Budget = {Total_budget}")
print(f"Average Math Score is  {Average_Math_score}")
print(f"Average Reading Score is  {Average_Reading_score}")
print(f"Passing Math % is  {passing_Math_count/Total_students*100}%")
print(f"Passing Reading % is  {passing_Reading_count/Total_students*100}%")
print(f"Passing Reading and Math % is  {passing_math_reading_count/Total_students*100}%")
```

```python
Per_School_Summary = district_summary.copy()
# add all columns created
total_students_per_school = Per_School_Summary.groupby('school_name')['Student ID'].count()

# Calculate total school budget per school
total_budget_per_school = Per_School_Summary.groupby('school_name')['budget'].first()

# Calculate average math and reading scores per school
average_math_score_per_school = Per_School_Summary.groupby('school_name')['math_score'].mean()
average_reading_score_per_school = Per_School_Summary.groupby('school_name')['reading_score'].mean()

# Calculate passing students for math and reading per school
passing_math_per_school = Per_School_Summary[Per_School_Summary['math_score'] >= 60].groupby('school_name')['Student ID'].count()
passing_reading_per_school = Per_School_Summary[Per_School_Summary['reading_score'] >= 70].groupby('school_name')['Student ID'].count()

# Calculate percentage passing math and reading per school
percentage_passing_math_per_school = (passing_math_per_school / total_students_per_school) * 100
percentage_passing_reading_per_school = (passing_reading_per_school / total_students_per_school) * 100

# Calculate overall passing percentage per school
percentage_overall_passing_per_school = (percentage_passing_math_per_school + percentage_passing_reading_per_school) / 2

# Calculate per student budget
per_student_budget = total_budget_per_school / total_students_per_school
```

```python
Per_School_Summary = pd.DataFrame({
    'School Type': schoolData.set_index('school_name')['type'],
    'Total Students': total_students_per_school,
    'Total School Budget': total_budget_per_school,
    'Per Student Budget': per_student_budget,
    'Average Math Score': average_math_score_per_school,
    'Average Reading Score': average_reading_score_per_school,
    '% Passing Math': percentage_passing_math_per_school,
    '% Passing Reading': percentage_passing_reading_per_school,
    '% Overall Passing': percentage_overall_passing_per_school})

print(Per_School_Summary)
```
